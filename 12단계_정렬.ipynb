```python
#통계학
import sys
print = sys.stdout.write
n = int(input())
num = [0 for i in range(8001)]
arr = []
mode = []
for i in range(n):
    n_in = int(sys.stdin.readline())
    num[4000 - n_in] += 1
    arr.append(n_in)
arr.sort()
m = max(num)
for i in range(8001):
    if m == num[i]:
        mode.append(4000 - i)
mode.sort()
print(str(round(sum(arr) / n)) + '\n')
print(str(arr[n // 2]) + '\n')
if len(mode) > 1:
    print(str(mode[1]) + '\n')
else:
    print(str(mode[0]) + '\n')
print(str(max(arr) - min(arr)))
```

```python
#소트인사이드
n = int(input())
list_n = []
for i in str(n):
    list_n.append(i)
list_n.sort(reverse=True)
for _ in list_n:
    print(_, end='')
```

```python
#좌표 정렬하기
import sys     
n = int(sys.stdin.readline())  
list_n = []       
for _ in range(n):
    x, y = map(int, sys.stdin.readline().split()) 
    list_n.append((x, y)) 
for i, j in sorted(list_n):  
    print(i, j)
```

```python
#좌표 정렬하기2
import sys
n = int(sys.stdin.readline())       
list_n = []           
for _ in range(n):        
    x, y = map(int, sys.stdin.readline().split())  
    list_n.append((y, x))   
for i, j in sorted(list_n): 
    print(j, i)        
```

```python
#단어 정렬
n = int(input())
list_n = []
for i in range(n):
    list_n.append(input())
temp = set(list_n)
list_n = list(temp)
list_n.sort()
list_n.sort(key = len)
for i in list_n:
    print(i)
```

```python
#나이순 정렬
n = int(input())
list_n = []
for i in range(n):
    list_n.append(list(input().split()))
list_n.sort(key= lambda x : int(x[0])) 
for i,j in list_n:
    print(i,j)
```
